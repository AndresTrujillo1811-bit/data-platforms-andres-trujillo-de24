```python
# b) Change the column names to get something similar to this:

import pandas as pd

excel = "data/riket2023_åk9_np.xlsx"
data_file = pd.read_excel(excel, sheet_name=None, skiprows=8)


# Renaming columns

def rename_columns(df):
    df.rename(columns={
        "Riket": "Plats", 
        "Typ av huvudman": "Huvudman", 
        "Flickor": "Flickor (A-F)", 
        "Pojkar": "Pojkar (A-F)",
        "Totalt": "Totalt (A-E)",
        "Flickor": "Flickor (A-E)",
        "Pojkar": "Pojkar (A-E)", 
        "Totalt": "Totalt (poäng)",
        "Flickor": "Flickor (poäng)", 
        "Pojkar": "Pojkar (poäng)"    
    }, inplace=True)
    return df

data_df = {sheet: rename_columns(df) for sheet, df in data_file.items()}
data_df

```

```python
# c) Do some initial exploratory data analysis on this dataset. 
# Use methods such as info, value_counts, columns, describe and some more to get a first insight into the dataset.

for sheet, df in data_file.items():
    print(f"\n {sheet} --- Information \n ---")
    print(df.info())
    print(f"\n {sheet} --- Description \n ---")
    print(df.describe())
```

```python
# d) Find out the amount of students that got F in mathematics (both girls and boys).

print("\n Students with F in math")
print(data_file.get("Matematik"))
```
